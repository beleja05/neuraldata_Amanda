```python
#Here I am importing the required libraries for this homework
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
#This puts the Sst and pvalb resting datasets into numpy arrays from their urls
Sst_url ='https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
Pvalb_url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest1 = np.loadtxt(Sst_url)
v_rest2 = np.loadtxt(Pvalb_url)
```

```python
#Stacks the two arrays on top of each other to make a (2,500) matrix
v_rest = np.vstack((v_rest1, v_rest2))
v_rest.shape
```

```python
#This checks how many values there are in v_rest
v_rest.size
```

```python
#This cell returns the first row of v_rest, which is the same as v_rest1
v_rest[0,:]
```

```python
#This adds together the values in the 3rd column
v_rest[:,2].sum()
```

```python
#This shows the value of the second to last value of the second row
v_rest[1,-2]
```

```python
#This creates a histogram of the Sst resting values.
plt.hist(v_rest[0,:])
plt.xlabel('Resting Potential')
plt.ylabel('Count')
```

```python
#This code finds the mean of both the pvalb and sst values
v_rest_mean_sst = v_rest[0,:].mean()
v_rest_mean_pvalb = v_rest[1,:].mean()
print(v_rest_mean_sst)
print(v_rest_mean_pvalb)
```

The resting potential for both Sst and Pvalb are relatively similar and around a typical neuron's resting value. Pvalb neurons are a bit more negitive a rest than Sst neurons are.

```python
#This imports scipy so we can do t-tests
import scipy.stats as stats
```

```python
#This runs a t-test to compare the
stats.ttest_ind(v_rest[0,:],v_rest[1,:])
```

This has a p-value of 7.1598e-09, so that means that the resting values are significantly different, since it is less than 0.05.
