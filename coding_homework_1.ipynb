```python
#This cell takes the urls for sst and pvalb datasets and reads them.
urlsst = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
pandas_data1 = pd.read_csv(urlsst)

urlpvalb = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
pandas_data2 = pd.read_csv(urlpvalb)
```

```python
#This code takes the previously read datasets and converted it into lists
data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
```

```python
#This flattens the lists to get rid of brackets around each element.
data1 = np.squeeze(data1)
data2 = np.squeeze(data2)
```
