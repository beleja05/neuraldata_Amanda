```python
#This imports the required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This put the data into np arrays
url1 ='https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_hits_hw5.csv'
url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_fa_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter=',')
acc_fa = np.loadtxt(url2, delimiter=',')
```

```python
#This shows the shape of each of the arrays
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#This creates a time variable between 0 and 4 seconds that has 8001 intervels
time = np.linspace(0, 4, 8001)
print(time.shape)
```

```python
#This calculates the sampling rate by dividing the number of samples by the total time
sampling_rate = 8001 / 4
print(sampling_rate)
```

```python
#This plots the first trial of acc_hits versus time on the left and first trial of acc_fa versus time on the right
fig, ax = plt.subplots(1, 2)
ax[0].plot(time, acc_hits[0, :])
ax[0].set_xlabel('Time')

ax[1].plot(time, acc_fa[0, :])
ax[1].set_xlabel('Time')

plt.show()
```

It seems like the dominant oscillation is about 3 Hz for the hits data and about 1 Hz in the fa data. The dominate seems to be a bit greater in hits than false alarms

```python
#This calculates the erp for acc_hits and then plots it with a center line and a line at where the screen touch occured
erp_acc_hits = np.mean(acc_hits, axis=0)

plt.plot(time, erp_acc_hits)
plt.axvline(x=2, color='black')
plt.axhline(y=0, color='black')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.show()
```

The voltage changes significantly after the screen touch, first having a sharp decrease, followed by a large increase. It seems like one relatively long wave cycle.

```python
#This calculates the erp for acc_fa and then plots it on the same graph as erp_acc_hits with a center line and a line at where the screen touch occured
erp_acc_fa = np.mean(acc_fa, axis=0)

plt.figure(figsize=(10, 6))
plt.plot(time, erp_acc_fa, color = 'blue')
plt.plot(time, erp_acc_hits, color ='red')
plt.axvline(x=2, color='black')
plt.axhline(y=0, color='black')
plt.xlabel('Time (s)')
plt.ylabel('Voltage')
plt.show()
```

The ERP changes after the screen touch in the hits group. In contrast, there is seemingly no difference in ERP for the false alarm group before and after screen touch.
