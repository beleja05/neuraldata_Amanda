```python
#This imports the needed libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#This cell imports the v_rest_sst and v_rest_pvalb datasets as np arrays
sst_url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
pvalb_url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_sst = np.loadtxt(sst_url)
v_rest_pvalb = np.loadtxt(pvalb_url)
```

```python
#This cell finds the mean of the v_rest_sst and v_rest_pvalb datasets. It then ran a t_test to compare the two. Then the code will print two histograms, one for each dataset, only if the p-value is < 0.05. Then it labels the y and x axis
print(np.mean(v_rest_sst))
print(np.mean(v_rest_pvalb))
t_test = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(t_test)
if t_test.pvalue < 0.05:
  fig, ax = plt.subplots(1,2)
  ax[0].hist(v_rest_sst)
  ax[1].hist(v_rest_pvalb)
  ax[0].set_xlabel('v_rest_sst')
  ax[1].set_xlabel('v_rest_pvalb')
  ax[0].set_ylabel('Frequency')
  ax[1].set_ylabel('Frequency')
  plt.show()
```

```python
#This takes the code from the pervious cell and make it into a function
def vrestHist():
  if t_test.pvalue < 0.05:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(v_rest_sst)
    ax[1].hist(v_rest_pvalb)
    ax[0].set_xlabel('v_rest_sst')
    ax[1].set_xlabel('v_rest_pvalb')
    ax[0].set_ylabel('Frequency')
    ax[1].set_ylabel('Frequency')
    plt.show()
```

```python
#This runs the function defined in the pervious cell
vrestHist()
```

```python
#This cell makes a scatterplot of the v_rest_sst and v_rest_pvalb values per row and colors them based on their values. It then labels the x and y axis.
listlen = len(v_rest_sst)
for i in range(listlen):
  if v_rest_sst[i] > -60 and v_rest_pvalb[i] > -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i],c='red')
  elif v_rest_sst[i] > -60 and v_rest_pvalb[i] < -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i],c='green')
  elif v_rest_sst[i] < -60 and v_rest_pvalb[i] < -60:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i],c='blue')
  else:
    plt.scatter(v_rest_sst[i], v_rest_pvalb[i],c='purple')
plt.xlabel('v_rest_sst')
plt.ylabel('v_rest_pvalb')
plt.show()
```
